```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('clustering_data.csv')
df.info()
df.head()
```

```python
plt.scatter(df['-5.462513918787130329e+00'],df['-8.113541922364099435e+00'] )
plt.show()
#from the scatter we can deduce that the number of clasters is between 3 to 5 
```

```python
#this is to get the optimum valiue of k
Sum_of_squared_distances = []
K = range(1,10)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(df)
    Sum_of_squared_distances.append(km.inertia_)

print(km)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
fig=plt.figure(figsize=(15,5))
plt.plot(K,Sum_of_squared_distances)
plt.grid(True)
plt.show()

#we can deduce the value of k from the graph when the Sum_of_squared_distances
#doesn't decrease significantly with every iteration
#we can clearly see that it's 3 (number of clusters is 3)
```

```python
x=df.iloc[:,[0,1]].values
kmeans=KMeans(n_clusters=3, init='random',random_state=1)
y_pred=kmeans.fit_predict(x)
plt.scatter(x[:,0],x[:,1],c=y_pred)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='red',label='centroids')
plt.show()
```
